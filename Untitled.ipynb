```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('train.csv')
```

```python
df = pd.DataFrame(data)
df
```

```python
da = df.drop(['Name', 'PassengerId', 'Ticket', 'Cabin'], axis=1)
da
```

```python
da['Age'].describe()
```

```python
da = da.fillna(df.mean())
da
```

```python
da.isnull().sum()
```

```python
d2 = da.dropna(subset=['Embarked'])
d2
```

```python
sns.set(style='whitegrid')
sns.stripplot(data=d2, x='Survived', y='Fare')
plt.show()
```

```python
sns.kdeplot(data=d2, x='Age', hue='Survived')
plt.show()
```

```python
sns.countplot(data=d2, x='Sex', hue='Survived')
```

```python
sns.countplot(data=d2, x='Pclass', hue='Survived')
plt.show()
```
